```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, MaxAbsScaler, RobustScaler

# Pisahkan kolom numerik dan non-numerik
non_numeric_cols = ['pekerjaan', 'label']
numeric_df = df[[col for col in df.columns if col not in non_numeric_cols]]

# Standardization
standard_scaler = StandardScaler()
standard_scaled_data = standard_scaler.fit_transform(numeric_df)
standard_scaled_df = pd.DataFrame(standard_scaled_data, columns=numeric_df.columns)
print("Standardization:\n", standard_scaled_df)

# Min-Max Scaling
minmax_scaler = MinMaxScaler()
minmax_scaled_data = minmax_scaler.fit_transform(numeric_df)
minmax_scaled_df = pd.DataFrame(minmax_scaled_data, columns=numeric_df.columns)
print("Min-Max Scaling:\n", minmax_scaled_df)

# MaxAbs Scaling
maxabs_scaler = MaxAbsScaler()
maxabs_scaled_data = maxabs_scaler.fit_transform(numeric_df)
maxabs_scaled_df = pd.DataFrame(maxabs_scaled_data, columns=numeric_df.columns)
print("MaxAbs Scaling:\n", maxabs_scaled_df)

# Robust Scaling
robust_scaler = RobustScaler()
robust_scaled_data = robust_scaler.fit_transform(numeric_df)
robust_scaled_df = pd.DataFrame(robust_scaled_data, columns=numeric_df.columns)
print("Robust Scaling:\n", robust_scaled_df)
```

```python
standard_scaled_df['label'] = df['label']
standard_scaled_df['pekerjaan'] = df['pekerjaan']
```

```python
from pycaret.classification import setup, compare_models, create_model, tune_model, evaluate_model, predict_model

# Memulai setup PyCaret
clf = setup(data=standard_scaled_df, target='label', verbose=False, html=False)

# Membandingkan model
best_model = compare_models()

# Membuat model
model = create_model(best_model)

# Men-tuning model
tuned_model = tune_model(model)

# Mengevaluasi model
evaluate_model(tuned_model)

# Prediksi menggunakan model yang sudah di-tuning
predictions = predict_model(tuned_model)
print(predictions)
```
